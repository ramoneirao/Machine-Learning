```python
import pandas as pd
pd.set_option('display.max_columns', 50)
df = pd.read_csv('traffic-collision-data-from-2010-to-present.csv')
```

```python
area_encode = pd.get_dummies(df['Area Name'])
```

```python
area_encode.head()
```

```python
concatenando = pd.concat([df, area_encode], axis=1)
concatenando.drop('Area Name', axis=1, inplace=True)
```

```python
concatenando.head()
```

```python
import matplotlib.pyplot as plt
concatenando.boxplot(column='Census Tracts')
plt.show()
```
